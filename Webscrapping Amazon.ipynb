```python
import requests
from bs4 import BeautifulSoup
import pandas

```

```python
r = requests.get("https://www.amazon.com/s?rh=n%3A16310101&page=2&qid=1595054330&ref=lp_16310101_pg_2", headers={'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'})
c=r.content
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
#print(soup.prettify())
```

```python
all=soup.find_all("span",{"class":"a-size-base-plus a-color-base a-text-normal"})
```

```python
len(all)
```
