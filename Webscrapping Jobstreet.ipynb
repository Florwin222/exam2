```python
l=[]
urlbase="https://www.jobstreet.com.ph/en/job-search/python-jobs/"
for page in range (1,50):
    url=(urlbase+str(page)+"/")
    print(urlbase+str(page)+"/")
    r = requests.get(url, headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})

    
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div",{"class":"FYwKg _17IyL_1 _2-ij9_1 _3Vcu7_1 MtsXR_1"})
    print (all)
    for item in all:
        d={}
        d["Position"]=item.find("div",{"class":"FYwKg _2j8fZ_1 sIMFL_1 _1JtWu_1"}).text
        try:
            d["Company"]=item.find("a",{"class":"qbDva _1weGM_1 _2CELK_1 FYwKg _2k9O2 _29sNS _58veS_1"}).text
        except:
            d["Company"]=None
        try:    
            d["Location"]=item.find_all("span",{"class":"FYwKg _1GAuD C6ZIU_1 _1PM5y_1 _3NFar_1 sQuda_1 _6ufcS_1 _2WTa0_1"})[0].text
        except:
            d["Location"]=None
        try:      
            d["Salary"]=item.find_all("span",{"class":"FYwKg _1GAuD C6ZIU_1 _1PM5y_1 _3NFar_1 sQuda_1 _6ufcS_1 _2WTa0_1"})[1].text
        except:
            d["Salary"]=None
        try:    
            d["Company Highlight"]=item.find("div",{"class":"FYwKg _1GAuD C6ZIU_1 _1PM5y_1 _3NFar_1 _29m7__1 _6ufcS_1 _2WTa0_1"}).text
        except:
            d["Company Highlight"]=None 


        d["Time of Posting"]=item.find("time",{"class":"FYwKg"}).text

        l.append(d)
   

```
